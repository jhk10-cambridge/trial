```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import figure
from ipywidgets import interact
from ipywidgets import interact, fixed, FloatSlider, HBox, VBox, Layout, Output, Label, Box, GridspecLayout, GridBox
import ipywidgets as widgets
from matplotlib import ticker


npts=100

y=np.linspace(0.0, 1.0, npts)
y2=np.linspace(0.0, 1.0, npts)

n_dict={'1s':1,'2s':2,'2p':2,'3s':3,'3p':3,'3d':3}
l_dict={'1s':0,'2s':0,'2p':1,'3s':0,'3p':1,'3d':2}

def genR(qn,ql,zeff,x,y,rdf):
  rho = 2.0 * zeff/qn
  if qn == 1 : 
    for i in range (0,npts): y[i] = 2.0 * zeff**1.5 * math.exp(-0.5 * rho*x[i])
  if qn == 2 and ql == 0: 
    for i in range (0,npts): y[i] = 0.353 * zeff**1.5 * (2.0 - rho*x[i]) * math.exp(-0.5 * rho*x[i])
  if qn == 2 and ql == 1: 
    for i in range (0,npts): y[i] = 0.204 * zeff**1.5 * rho*x[i] * math.exp(-0.5 * rho*x[i])
  if qn == 3 and ql == 0: 
    for i in range (0,npts): y[i] = 0.06415 * zeff**1.5 * (6.0 - 6.0* rho*x[i]+(rho*x[i])**2) * math.exp(-0.5 * rho*x[i])
  if qn == 3 and ql == 1: 
    for i in range (0,npts): y[i] = 0.04536 * zeff**1.5 * (4.0 - rho*x[i])*rho*x[i] * math.exp(-0.5 * rho*x[i])
  if qn == 3 and ql == 2: 
    for i in range (0,npts): y[i] = 0.02029 * zeff**1.5 * (rho*x[i])**2 * math.exp(-0.5 * rho*x[i])
  if rdf:
    for i in range (0,npts): y[i] = x[i]**2 * y[i]**2



def makeplot(rmax,ymin,ymax,sel_orb_1,pl_1,zeff_1,sel_orb_2,zeff_2,pl_2):
# sel_n1.observe(handle_slider_change, names='value')
  qn=n_dict[sel_orb_1]
  ql=l_dict[sel_orb_1]  
  x = np.linspace(0, rmax, npts)
  rdf=False
  if pl_1=='plot RDF': rdf=True
  genR(qn,ql,zeff_1,x,y,rdf)
  if pl_2:
    qn=n_dict[sel_orb_2]
    ql=l_dict[sel_orb_2]  
    x2 = np.linspace(0, rmax, npts)
    genR(qn,ql,zeff_2,x2,y2,rdf)
#designed for figsize=(8,4.94) {Golden ratio, 1.618}
  fig = plt.figure(figsize=(8,8/1.618)) 
  ax = fig.add_subplot(111)
  plt.plot(x,y,linewidth=2,linestyle='-',color='red',clip_on=True)
  if pl_2: plt.plot(x2,y2,linewidth=2,linestyle='-',color='blue',clip_on=True)
#  plt.axis([0,rmax,ymin,ymax])
  if pl_1=='RDF(r)':
    plt.ylabel(r'$\rho(r)$',fontsize=12)
  else :
    plt.ylabel(r'$R(r)$',fontsize=12)
  plt.xlabel(r'$r/a_0$',fontsize=12)
  x_ticks = np.arange(0,rmax+1.0,1.0)
  if ymax/0.01 >= 3 : ystep = 0.01
  if ymax/0.02 >= 3 : ystep = 0.02
  if ymax/0.05 >= 3 : ystep = 0.05
  if ymax/0.1 >= 3 : ystep = 0.1
  if ymax/0.2 >= 3 : ystep = 0.2
  if ymax/0.5 >= 3 : ystep = 0.5
  ytop = math.ceil(ymax/ystep) * ystep
  ymin_loc=ymin
  if pl_1=='RDF(r)': ymin_loc=0.0
  if abs(ymin_loc) < 0.001 :
    ybot = 0.0
  else :
    ybot = - math.ceil(abs(ymin_loc)/ystep) * ystep
  y_ticks = np.arange(ybot, ytop+ystep, ystep)
  if ybot < 0.0 : plt.plot([0,rmax],[0.0,0.0],linewidth=1,linestyle='-',color='black')
  plt.xticks(x_ticks)
  plt.yticks(y_ticks)
  plt.axis([0,rmax,ymin_loc,ymax])
  plt.grid(True,color='blue',linewidth=.25)
#  ax.set_xticklabels([])
#  ax.set_yticklabels([])
  plt.tick_params(axis='x',which='both',bottom = True, top= True)
  plt.tick_params(axis='y',which='both',left = True, right= True)
  plt.show()

#margin='top right bottom left'
ymx=FloatSlider(description='',min=.05, max=2, step=0.01, value=2,continuous_update=True,
orientation='vertical',readout=False,
style=dict(handle_color='lightcoral'),
layout=dict(font_color='red',height='100px',width='auto',margin='0px 0px 0px 0px'))
#300% is to push the bottom of the slider up


#margin='top right bottom left'
ymn=FloatSlider(description='',min=-0.2, max=0, step=0.01, value=-0.1,continuous_update=True,
orientation='vertical',readout=False,
style=dict(handle_color='lightcoral'),
layout=dict(height='100px',width='auto',margin='0px 0px 0px 0px'))
#300% is to push the bottom of the slider up

#rm=FloatSlider(description='rm',min=0, max=12, step=1, value=10,continuous_update=True,
#orientation='vertical',readout=False,
#style=dict(handle_color='lightblue'),
#layout=dict(width='auto', height='auto',margin='0px 0px 300% 0px'))

rm=FloatSlider(description='',min=2.0, max=25.0, step=1, value=15.0,continuous_update=True,
orientation='horizontal',readout=True,readout_format='.0f',
style=dict(description_width='60px',handle_color='gray'),
layout=dict(width='400px', height='25px',margin='0px 0px 0px 0px'))

sel_orb_1=widgets.Dropdown(options=['1s','2s','2p','3s','3p','3d'],value='2s',description='orbital',
disabled=False,style=dict(description_width='60px'),
layout=dict(width='120px',margin='0px 0px 0px 0px'))

sel_orb_2=widgets.Dropdown(options=['1s','2s','2p','3s','3p','3d'],value='2p',description='orbital',
disabled=False,style=dict(description_width='60px'),
layout=dict(width='120px',margin='0px 0px 0px 0px'))

zeff_1=FloatSlider(description='',min=1, max=5.0, step=.11, value=1.0,continuous_update=True,
orientation='horizontal',readout=True,readout_format='.1f',
style=dict(description_width='25px',handle_color='gray'),
layout=dict(width='200px', height='30px',margin='0px 0px 0px 0px'))

zeff_2=FloatSlider(description='',min=1, max=5.0, step=.11, value=1.0,continuous_update=True,
orientation='horizontal',readout=True,readout_format='.1f',
style=dict(description_width='25px',handle_color='gray'),
layout=dict(width='200px', height='30px',margin='0px 0px 0px 0px'))

#sel_n1=widgets.Dropdown(options=[1,2,3],value=2,description='n',
#disabled=False,style=dict(description_width='20px'),
#layout=dict(width='auto',margin='0px 0px 0px 0px'))

#lopts=[0,1,2]

def handle_slider_change(change):
    if change.new==1 : lopts=[0]
    if change.new==2 : lopts=[0,1]
    if change.new==3 : lopts=[0,1,2]

#sel_n1.observe(handle_slider_change, names='value')

#sel_l1=widgets.Dropdown(options=lopts,value=0,description='l',
#disabled=False,style=dict(description_width='20px'),
#layout=dict(width='auto',margin='0px 0px 0px 0px'))



pl_1=widgets.RadioButtons(options=['plot radial part', 'plot RDF'],
value='plot radial part', description='',disabled=False,
layout=dict(width='auto',margin='0px 0px 0px 0px'))

pl_2=widgets.Checkbox(value=False,description='plot second function?',disabled=False,indent=False,
layout=dict(margin='40px 0px 10px 0px',width='150px'))

#out = widgets.interactive_output(pvplot, {'a': a, 'b': b, 'xsl':xsl})
out = widgets.interactive_output(makeplot, 
{'rmax': rm, 'ymin': ymn, 'ymax': ymx, 'sel_orb_1':sel_orb_1,'pl_1':pl_1,'zeff_1':zeff_1,
 'sel_orb_2':sel_orb_2,'zeff_2':zeff_2,'pl_2':pl_2})

zeff_label=widgets.HTMLMath(value='<i>Z</i><sub>eff</sub>')
max_label=widgets.HTMLMath(value='<p style="font-size:10px">max <i>y</i></p>',layout=dict(margin='0px 0px 0px 0px'))
min_label=widgets.HTMLMath(value='<p style="font-size:10px">min <i>y</i></p>',layout=dict(margin='0px 0px 0px 0px'))
rmax_label=widgets.HTMLMath(value='<p style="font-size:10px"><i>r</i><sub>max</sub>/<i>a</i><sub>0</sub></p>',layout=dict(margin='0px 0px 0px 50px'))


#the height of the top row is set by the height of the graph
#the slides are 'auto' height, but with a large margin at the bottom
#to place them correctly
GridBox(children=[VBox([max_label,ymx,min_label,ymn]),out,VBox([sel_orb_1,pl_1,HBox([zeff_label,zeff_1]),pl_2,sel_orb_2,HBox([zeff_label,zeff_2])]),
                  HBox([]),HBox([rmax_label,rm]),HBox([])],
        layout=Layout(
            width='100%',
            grid_template_columns='3% 40% 20%',
            grid_template_rows='auto auto',
            grid_gap='0px 0px')
       )

```
